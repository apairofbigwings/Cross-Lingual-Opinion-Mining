# Get topwords only by tf-idf

```
directory = '/content/drive/My Drive/Colab Notebooks/opinion-lab-group-2.3/refactorization/data/kmean-result-backup/'

def get_topwords(k: int, sentences: np.array, labels: np.array, stopwords: topwords.Stopwords):
  """
  k: int
  	number of clusters
  sentences: 1d np.array
    an array of sentences (of a language or a source)
  labels: 1d np.array
    an array of cluster labels corresponding to sentences
  stopwords: frozenset
    choice in Stopwords class
  use_global_vectorizer: boolean
    if True, use global idf (idf calculated from all sentences of corpus, i.e. all sentences from a language or a source)
    if False, use cluster idf (idf calculated from all sentences labelled as same cluster in the corpus)
  """
  # print('Global')
  vocabulary, bag_of_words, tfidf, global_vectorizer = topwords.get_tfidf(stopwords, sentences)

  tfidf_k = []
  for i in range(k):
    # print('Cluster', i)
    sentences_i = sentences[np.where(labels == i)]

    # if use_global_vectorizer:
    #   _, _, t, _ = topwords.get_tfidf(stopwords, sentences_i, vocabulary = vocabulary, global_vectorizer = global_vectorizer)
    # else:
    _, _, t, _ = topwords.get_tfidf(stopwords, sentences_i, vocabulary = vocabulary)
    tfidf_k.append(t)

  tfidf_k = np.array(tfidf_k)
  # clarity_score = np.apply_along_axis(get_clarity_score, 1, tfidf_k, y = tfidf) # a row indicates tfidf for a cluster
  
  sorted_scores = np.apply_along_axis(topwords.get_sort_scores, 1, tfidf_k)
  sorted_indeces = np.apply_along_axis(topwords.get_sort_indeces, 1, tfidf_k)
  sorted_words = np.array(bag_of_words)[sorted_indeces]

  print('Top words and their corresponding tf-idf for k = {} are ready!'.format(k))
  return sorted_words, sorted_scores


def get_result(k, directory = directory):
  _, labels = load_kmean_labels(directory + str(k) + '_kmean_labels.npy')

  en_labels = labels[:len(en_sentences)]
  de_labels = labels[len(en_sentences):len(en_sentences) + len(de_sentences)]
  labels = None

  assert en_labels.shape[0] == en_sentences.shape[0]
  assert de_labels.shape[0] == de_sentences.shape[0]
  
  print()
  print()
  en_stemmed_sentences = get_stemmed_sentences(Language.EN, Stopwords.EN_TFHIGH, en_sentences)
  en_top_words, en_top_scores = get_topwords(k, en_stemmed_sentences, en_labels, Stopwords.EN_TFHIGH)
  
  print()
  print()
  de_stemmed_sentences = get_stemmed_sentences(Language.DE, Stopwords.DE_TFHIGH, de_sentences)
  de_top_words, de_top_scores = get_topwords(k, de_stemmed_sentences, de_labels, Stopwords.DE_TFHIGH)
  # topwords.print_wordcloud(k, de_top_words, de_top_scores, title = 'Word cloud for German corpus with stemmed words')

  return en_top_words, en_top_scores, de_top_words, de_top_scores
```

```
def get_cosine_similarity(x, y):
  """
  x, y: 1d array of same length
    array of tdidf scores of a cluster and global tdidf scores respectively
  """
  return np.dot(np.squeeze(x), np.squeeze(y))

for de_embedding in de_embeddings:
  cs = np.apply_along_axis(get_cosine_similarity, 1, en_embeddings[0:10], y = de_embedding)

cs
```

```
result = []
count = 5

for en_index, en_embedding in enumerate(en_embeddings[0:10]):
  tuples = []
  for de_index, de_embedding in enumerate(de_embeddings):
    similarity_score = np.dot(np.squeeze(en_embedding), np.squeeze(de_embedding))
    de_index = en_embeddings.shape[0] + de_index
    tuples.append((en_index, de_index, similarity_score))

    result.append(pd.DataFrame(tuples, columns = ['en_index', 'de_index', 'similarity_score']).sort_values('similarity_score', ascending = False).head(count).to_numpy())
```
